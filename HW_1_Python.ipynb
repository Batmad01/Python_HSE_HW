## Анализ исторических данных

```python
data = pd.read_csv('temperature_data.csv')
display(data.head(5))
data.info()
```

```python
# Меняем тип данных на datetime
data['timestamp'] = pd.to_datetime(data['timestamp'])
data = data.sort_values(by=['city', 'timestamp'])
```

```python
# Функция для расчета скользящего среднего за 30 дней
def moving_avg(group):
    group['moving_average'] = group.groupby('city')['temperature'].transform(
        lambda x: x.rolling(window=30, min_periods=1).mean())
    return group

# Функция для вычисления сезонных статистик и выделения аномалий
def season_stats(group):
    stats = group.groupby(['season', 'city'])['temperature'].agg(['mean', 'std']).reset_index()
    group =  group.merge(stats, on=['season', 'city'], suffixes=('', '_seasonal'))
    group['anomaly'] = abs(group['temperature'] - group['mean']) > 2 * group['std']
    return group

# Распараллеливание обработки
def par_process(group):
    group = moving_avg(group)
    group = season_stats(group)
    return group

# Разделение данных по городам
data_groups = [group for _, group in data.groupby('city')]
```

```python
# Запустим несколько раз, чтобы поймать различия
for i in range(3):
    # Расчет времени выполнения без распараллеливания
    start_time = time.time()
    data = pd.concat([moving_avg(group) for group in data_groups], ignore_index=True)
    data = pd.concat([season_stats(group) for group in data_groups], ignore_index=True)
    no_par_time = time.time() - start_time

    # С распараллеливанием
    start_time = time.time()
    with ThreadPoolExecutor(max_workers=4) as executor:
        results = list(executor.map(par_process, data_groups))
    final_df = pd.concat(results, ignore_index=True)
    par_time = time.time() - start_time

    print(f"Время выполнения без распараллеливания: {no_par_time:.2f} секунд")
    print(f"Время выполнения с распараллеливанием: {par_time:.2f} секунд")
    print()
```

`С распареллеливанием в среднем немного медленее`

```python
# Для графика
season_order = ['autumn', 'winter', 'spring', 'summer']
data['season'] = pd.Categorical(data['season'], categories=season_order, ordered=True)
seasonal_stats = data.groupby(['season', 'city'], observed=False)['temperature'].agg(['mean', 'std']).reset_index()

# График средних значений температуры для каждого сезона в каждом городе
plt.figure(figsize=(12, 6))
for city in seasonal_stats['city'].unique():
    city_data = seasonal_stats[seasonal_stats['city'] == city]
    plt.plot(city_data['season'], city_data['mean'], marker='o', label=city)

plt.title('Средняя температура по сезонам в разных городах')
plt.xlabel('Сезон')
plt.ylabel('Средняя температура (°C)')
plt.legend(title='Город', loc='lower right')
plt.grid(True)
plt.show()
```

```python
# График аномалий температуры для Москвы
sample_city = 'Moscow'
city_anomalies = data[(data['city'] == sample_city) & (data['anomaly'])]

plt.figure(figsize=(12, 6))
plt.plot(data[data['city'] == sample_city]['timestamp'], 
         data[data['city'] == sample_city]['moving_average'], label='Температура')
plt.scatter(city_anomalies['timestamp'], 
            city_anomalies['moving_average'], color='green', label='Аномалия', zorder=2)
plt.title(f'Температура и аномалии в Москве')
plt.xlabel('Год')
plt.ylabel('Температура (°C)')
plt.legend()
plt.grid(True)
plt.show()
```

## Мониторинг текущей температуры

```python
# API ключ
with open('key.txt') as file:
    api_k = file.read().strip()

# Какие параметры для города будем забирать из API
params = ['country', 'lat', 'lon']

# Синхронная функция получения местоположения города и температуры в нём
# На момент написания кода студент (я) ещё не знал, что можно получить данные
# о городе с помощью всего одного запроса к API, поэтому тут чуть сложнее получилось

def city_data(data, api_key):
    # Данные о городах (страна, широта, долгота)
    cities_dict = {}
    for city in data['city'].unique():
        response = requests.get(f'http://api.openweathermap.org/geo/1.0/direct?q={city},&limit={1}&appid={api_key}')
        cities_dict[city] = [response.json()[0][x] for x in params]

    # Температуры по городам
    temp_now = {}
    for city in cities_dict:
        lat, lon = cities_dict[city][-2], cities_dict[city][-1]
        response = requests.get(f'https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}&units=metric')
        temp_now[city] =  response.json()['main']['temp']

    return cities_dict, temp_now
```

```python
# Асинхронная функция получения местоположения города и температуры в нём
async def city_data_async(data, api_key):
    cities_dict = {}

    # Данные о городах (страна, широта, долгота)
    async def fetch_city(city):
        async with aiohttp.ClientSession() as session:
            async with session.get(f'http://api.openweathermap.org/geo/1.0/direct?q={city},&limit={1}&appid={api_key}') as response:
                json_response = await response.json()
                return city, [json_response[0][x] for x in params]
    
    # Температуры по городам
    temp_now = {}
    async def fetch_temp(city, lat, lon):
        async with aiohttp.ClientSession() as session:
            async with session.get(f'https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}&units=metric') as response:
                json_response = await response.json()
                return city, json_response['main']['temp']
    
    # Запускаем таски по сбору информации по городам
    city_tasks = [fetch_city(city) for city in data['city'].unique()]
    city_results = await asyncio.gather(*city_tasks)

    for city, city_data in city_results:
        cities_dict[city] = city_data

    # Запускаем таски по сбору температур по городам
    temp_tasks = [fetch_temp(city, cities_dict[city][-2], cities_dict[city][-1]) for city in cities_dict]
    temp_results = await asyncio.gather(*temp_tasks)
    
    temp_now = {city: temp for city, temp in temp_results}

    return(cities_dict, temp_now)
```

```python
# Синхронное выполнение
start_time = time.time()
sync_res, sync_temp = city_data(data, api_k)
sync_time = time.time() - start_time
print(f"Синхронное выполнение: {sync_time:.2f} секунд")

# Асинхронное выполнение
start_time = time.time()
async_res, async_temp = asyncio.run(city_data_async(data, api_k))
async_time = time.time() - start_time
print(f"Асинхронное выполнение: {async_time:.2f} секунд")
print(f'Равенство полученных результатов: {sync_res == async_res}')
```
